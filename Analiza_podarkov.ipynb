## Kakšne zvrsti je serija

```python
def koren_besede(beseda):
    beseda = ''.join(znak for znak in beseda if znak.isalpha())
    if not beseda:
        return '$'
    return beseda.rstrip('ds').rstrip('aeiou')

def koreni_besed(niz):
    return pd.Series(sorted({koren_besede(beseda) for beseda in niz.replace('-', ' ').lower().split()}))

serije
```

```python
filmi = serije[serije['leto-konec'] - serije['leto-zacetek'] == 0]
filmi
```

```python
zanri
```

```python

st_filmov_po_zanru = zanri.groupby('zanr').size()
st_vseh_filmov = len(filmi)
st_vseh_filmov
```

* Ugotovimo verjetnost filmov

```python

verjetnost_zanra = st_filmov_po_zanru / st_vseh_filmov
verjetnost_zanra
```

```python
opisi_in_zanri = pd.merge(filmi, zanri, left_on='id', right_on='film')[['zanr', 'opis']]
```

*Dobili smo opis filmov razdeljene v korene

```python
opisi_in_zanri
```

```python
tabela_korenov_in_filmov = filmi.opis.apply(koreni_besed)
```

```python
tabela_korenov_in_filmov
```

Komentar profesorju: kodo sem želel sam spisat, a nikakor nisem mogel prenest id ja filma k razdelitvi žanrov in filmov, ker se število filmov podvoji-ali potroji ko združimo žanre in filme. tako da sem vzel vašo, kar se tiče branja iz 'id'. 

```python
koreni_in_filmi = tabela_korenov_in_filmov.stack().reset_index()[['id', 0]].rename(columns={'id': 'film', 0: 'koren'})
```

```python
koreni_in_filmi
```

```python
koreni_in_zanri = pd.merge(koreni_in_filmi, zanri)
koreni_in_zanri
```

```python
stevila_korenov_pri_zanru = pd.crosstab(koreni_in_zanri.koren, koreni_in_zanri.zanr)
```

```python
stevila_korenov_pri_zanru
```

```python
st_filmov_po_zanru
```

```python
verjetnosti_korenov_pri_zanru = (stevila_korenov_pri_zanru + 0.1) / st_filmov_po_zanru
```
